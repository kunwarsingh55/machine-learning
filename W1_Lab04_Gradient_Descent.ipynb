# Function to calculate cost, J(w,b)
#### - Squared Error Function
#### - Model function : fw,b(x(i)) = (w*x(i) + b)
#### - J(w,b) = 1/2m (i=i to i = m) Σ(w*x(i)+y(i))^2



```python
#Compute J(w,b) function
#Computes cost with new w b for all x and y
def compute_cost(x, y, w, b):
    cost = 0
    summition = 0
    m = x.shape[0]
    for i in range(m):
        f_wb = w*x[i] + b
        summition = summition + ((f_wb-y[i])**2)
    cost = (1/2*m) + summition
    return cost

```

# Implement Gradient Descent
### The algorithm
    - Repeate until convergenct{
        w = w - alpha(d/dw)J(w,b)
        b = b - alpha(d/db)J(w,b)
    }
    
    (d/dw)J(w,b) = 1/m * (i=i to i = m) Σ(w*x(i)+y(i))x(i)
    (d/db)J(w,b) = 1/m * (i=i to i = m) Σ(w*x(i)+y(i))
    

```python
#Function to compute gradient at a particular point.
# Just the gradient(slop/deravitives) not the dencent...
def compute_gradient(x, y, w, b):
    m = x.shape[0]
    dj_dw = 0  #Derivative of J(w,b) w.r.t w
    dj_db = 0  #Deravitive of J(w,b) w.r.t b
    dw_summition = 0
    db_summition = 0
    for i in range(m):
        f_wb = w*x[i] + b
        dw_summition = dw_summition + (f_wb - y[i]) * x[i]
        print(x[i])
        db_summition = db_summition + (f_wb - y[i])
    dj_dw = dw_summition / m
    dj_db = db_summition / m
    
    return dj_dw, dj_db
    
    
```

```python
#Main Gradient Decent
def gradient_descent(x, y, alpha, num_iterate):
    # GD starts with some val of w and b
    # Let w and b = 0
    # alpha = learning rate
    # num_iterate = num of gradient descent iterations to perform
    
    w = 0
    b = 0
    cost_history = []
    for i in range(num_iterate):
        dj_dw, dj_db = compute_gradient(x, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        cost_history.append(compute_cost(x, y, w, b))
    print(w , b)
    v = 0
    ite_axis = []
    hist = []
    for i in range(num_iterate):
        if v == 1000:
            #print(cost_history[i])
            v = 0 
            hist.append(cost_history[i]*1000)
            ite_axis.append(i)
        v = v + 1
        
       
    plt.plot(hist, ite_axis , c='b',label='Our Prediction')
    
```
